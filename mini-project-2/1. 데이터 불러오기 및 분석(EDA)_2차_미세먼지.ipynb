#### [실습문제4] 미세먼지 데이터 알아보기
* air_21,air_22 의 다양한 변수에 대하여 여러 그래프를 그려보고 인사이트를 도출해보세요.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
# 지역	망	측정소코드	측정소명 주소 => 범주
# 측정일시	=> 시간
#SO2	CO	O3	NO2	PM10	PM25 => 수치
cols = air_21.columns
index = 0
plt.figure(figsize=(16,18))
for col in cols:
    index += 1
    plt.subplot(6,2,index)
    sns.histplot(air_21[col], kde=True)
plt.tight_layout()
plt.show()
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
cols = ['지역', '망', '측정소코드', '측정소명', '주소']
index = 0
plt.figure(figsize=(16,8))
for col in cols:
    index += 1
    plt.subplot(3,2,index)
    temp = air_21[col].value_counts()
    plt.pie(temp.values, labels=temp.index, shadow=True)
plt.show()


```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
plt.figure(figsize=(16,4))
sns.heatmap(air_21.corr())
plt.show()
```

```python
# 위 차트를 통해 알게된 사실을 정리해봅시다.
# 1. 지역, 망, 측정소코드, 측정소명, 주소는 단 1개 값만 있다.
# 2. PM10 과 PM2.5는 연관이 있어보인다.
# 3. NO2와 CO는 연관이 있어 보인다. SO2 CO O3 NO2 = 아황산가스	일산화탄소	오존	이산화질소
# 4. CO와 O3는 연관이 없어 보인다.
# 5. 측정소 코드는 연관이 없다.
```
